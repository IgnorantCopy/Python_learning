# 打开文件

open参数：
1. file:用来指定打开文件（文件的路径）
2. mode:打开文件的模式，默认是 r 
    + r:只读（打开文件以后，只能读取，不能写入），如果文件不存在，会报错
    + w:只写（打开文件以后，只能写入，不能读取），如果文件存在，会覆盖文件；如果文件不存在，会创建文件
    + b:以二进制的形式打开文件，可以用来操作非文本文件
    + rb:以二进制的形式读取文件   wb:以二进制的写入文件
    + a:追加（在最后追加内容），如果文件不存在，会创建文件
    + r+/w+:可读写， r+ 文件不存在，会报错, w+ 文件存在，会覆盖文件；文件不存在，会创建文件（不建议使用）
3. encoding:打开文件的编码方式

```python
file = open('xxx', 'r')  # gbk
print(file.read())
file.close()  # 操作完后，关闭文件
```

```python
file = open('abc', 'r', encoding='utf8')  # abc使用utf8编码格式，windows默认使用gbk编码格式
print(file.read())
file.close()
```

```python
file = open('xxx', 'w')
file.write('hello')  # 只能写入字符串
file.close()
```

```python
file = open('abc', 'rb')
print(file.read())  # 读取结果是二进制
file.close()
```

```python
file = open('abc', 'wb')
file.write('你好'.encode('utf8'))
file.close()
```

# 文件的拷贝

```python
import os

file_name = input('请输入一个文件路径：')
if os.path.isfile(file_name):
    old_file = open(file_name, 'rb')
    new_file = open(file_name + '.bak', 'wb')  #names=file_name.rpartition('.')
    while 1:  #new_name=names[0]+'.bak.'+names[2]
        content = old_file.read(1024)
        new_file.write(content)
        if not content:
            break
    new_file.close()
    old_file.close()
else:
    print('文件不存在')
```

# 将数据写入内存
* 涉及 StringIO 和 BytesIO 两个类

```python
from io import StringIO

s_io = StringIO()
s_io.write('hello')  #把数据写入到内存里缓存起来了
print(s_io.getvalue())
s_io.close()
```

```python
from io import StringIO

s_io = StringIO()
print('world', file=s_io)
print(s_io.getvalue())
s_io.close()
```

```python
from io import BytesIO

b_io = BytesIO()
b_io.write('hello'.encode('utf8'))
print(b_io.getvalue().decode('utf8'))
b_io.close()
```

# json进阶

## 序列化：将数据从内存持久化保存到硬盘的过程

```python
#将字典、列表、数字等写入文件
#1.将数据转换成字符串或二进制
#2.使用json模块
import json

name = ['zhangsan', 'lisi', 'wuangwu']
x = json.dumps(name)  #将数据转换成json字符串，不会将数据保存到文件里
file = open('abc', 'w')
file.write(x)
file.close()
```

```python
import json

name = ['zhangsan', 'lisi', 'wuangwu']
file = open('abc', 'w')
x = json.dump(name, file)  # 将数据转换成json字符串，同时写入到指定文件
file.close()
```

# try...except语句
* 用来处理运行过程中的异常

```python
def div(a, b):
    return a / b


try:
    x = div(5, 0)
    print('hello')  #不会运行
except Exception as e:  #如果程序出错，会立即跳转到 except 语句
    print('程序出错！')
else:
    print('计算的结果是', x)
```

```python
def div(a, b):
    return a / b


try:
    x = div(5, 2)
    print('hello')
except Exception as e:
    print('程序出错！')
else:
    print('计算的结果是', x)
```

```python
try:
    file = open('xxxx')
    print(file.read())
    file.close()
except Exception as e:  #异常变量名:e  异常类型:Exception
    print('程序出错')
    print(e)
```

```python
try:
    file = open('xxxx')
    print(file.read())
    file.close()
    print(1 / 0)
except (FileNotFoundError, ZeroDivisionError) as e:  #异常变量名:e  异常类型:Exception
    print('程序出错')
```

## finally关键字的使用

```python
file = open('素材/开始.jpg', 'rb')
try:
    while 1:
        content = file.read(64)
        if not content:
            break
        print(content)
finally:  # 最终都会被执行的代码
    print('文件被关闭了')
    file.close()
```

### 注意事项

```python
def demo(a, b):
    try:
        x = a / b
    except ZeroDivisionError:
        return '除数不能为零'
    else:
        return x
    finally:
        return 'good'  #finally里的返回值会覆盖之前的返回值


print(demo(1, 0))
```

# with关键字的使用
* 上下文管理器，很多需要手动关闭的连接，如：文件连接、socket连接、数据库的连接，都能使用 with 关键自动关闭连接

```python
#try:
#    file=open('xxx','r')
#except FileNotFoundError:
#    print('文件不存在')
#else:
#    try:
#        file.read()
#    finally:
#        file.close()
try:
    with open('xxx', 'r') as file:  #with 关键字会帮助关闭文件
        file.read()
except FileNotFoundError:
    print('文件未找到')
#with 关键字后面对象，需要实现 __enter__ 和 __exit__ 魔法方法
```

## 上下文管理器

```python
#with 语句后面的结果对象，需要重写 __enter__ 和 __exit__ 方法
class Demo(object):
    def __enter__(self):
        print('__enter__被调用了')
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print('__exit__被调用了')


def test():
    return Demo()


with test() as t:
    print(t)  #t==>test.__enter__()
```

# 自定义异常

```python
#让用户输入6~12位的密码
class LenError(Exception):
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return '长度必须在{}到{}位之间'.format(self.x, self.y)


password = input('请输入密码：')
if 6 <= len(password) <= 12:
    print('密码正确')
else:
    #使用 raise 关键字可以抛出一个异常
    raise LenError(6, 12)

print('将密码保存到数据库')
```
